```python
import numpy as np  # X_huge 생성에 필요한 numpy 임포트 추가
from sklearn.decomposition import PCA, IncrementalPCA  # PCA 임포트 추가
from sklearn.metrics.pairwise import cosine_similarity

# 1. 데이터 생성 (10,000 샘플, 50차원)
np.random.seed(42)  # 일관된 결과를 위해 시드 추가
X_huge = np.random.randn(10000, 50)

# 2. 일반 PCA 실행
pca = PCA(n_components=5)
X_pca = pca.fit_transform(X_huge)

# 3. Incremental PCA 실행 (미니배치 1,000개씩 공분산 업데이트)
ipca = IncrementalPCA(n_components=5, batch_size=1000)
X_ipca = ipca.fit_transform(X_huge)

# 4. 두 모델의 주성분 축(Components) 간의 코사인 유사도 측정
cos_sim = cosine_similarity(pca.components_, ipca.components_)

print("=== PCA vs Incremental PCA 주성분 축 코사인 유사도 ===")
for i in range(5):
    # 대각 성분(동일 주성분 간 비교) 확인
    print(f"주성분 {i+1}번 유사도: {abs(cos_sim[i, i]):.6f}")
```
