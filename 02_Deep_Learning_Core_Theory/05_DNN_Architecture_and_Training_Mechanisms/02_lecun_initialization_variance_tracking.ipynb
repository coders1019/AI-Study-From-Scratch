```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt

def run_init_experiment(init_type='lecun', layers=10, dim=500):
    # 입력 데이터의 분산을 1로 설정
    X = torch.randn(1000, dim)

    layer_outputs = []
    current_input = X

    for i in range(layers):
        linear = nn.Linear(dim, dim, bias=False)

        # 초기화 전략 적용
        if init_type == 'bad_normal':
            # 가중치 분산 조절 안 함 (표준편차 1)
            nn.init.normal_(linear.weight, mean=0.0, std=1.0)
        elif init_type == 'lecun':
            # LeCun 초기화: Var = 1 / fan_in -> std = sqrt(1 / fan_in)
            fan_in = dim
            nn.init.normal_(linear.weight, mean=0.0, std=np.sqrt(1.0 / fan_in))

        current_input = linear(current_input)
        layer_outputs.append(current_input.var().item())

    return layer_outputs

# 두 가지 케이스 실행
bad_vars = run_init_experiment('bad_normal')
lecun_vars = run_init_experiment('lecun')

# 결과 시각화
plt.figure(figsize=(10, 4))
plt.plot(bad_vars, label='Random Normal (No Scaling)', marker='o')
plt.plot(lecun_vars, label='LeCun Initialization', marker='s')
plt.yscale('log') # 값이 폭발하므로 로그 스케일로 관찰
plt.xlabel('Layer Depth')
plt.ylabel('Variance of Output (Log Scale)')
plt.title('Variance Explosion vs Preservation')
plt.legend()
plt.show()
```
